```python
import requests
```

```python
req = requests.get('https://api.covid19api.com/all')
```

```python
data = req.json()
```

```python
print(len(data))
```

```python
indonesia = 0
for i in range(len(data)):
    if (data[i]['Country']=='Indonesia'):
        indonesia = indonesia+1
        print(data[i])
```

```python
print(indonesia)
```

```python
import pandas as pd
```

```python
dunia = pd.DataFrame(data)
```

```python
indonesia = dunia[dunia['Country']=='Indonesia']
```

```python
indonesia.drop(['Province', 'City','CityCode'], axis = 1, inplace=True)
```

```python
indonesia
```

```python
indonesia.Date = pd.to_datetime(indonesia['Date']).dt.date
```

```python
indonesia.to_csv('Data_Covid19_Indonesia.csv', index=False)
```
